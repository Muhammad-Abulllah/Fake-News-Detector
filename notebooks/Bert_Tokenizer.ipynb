```python
from transformers import BertTokenizer, BertModel
import torch
```

```python
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
print("Vocabulary size : ", tokenizer.vocab_size)
```

```python
sentence = "This news article is completely fake and misleading."
tokens = tokenizer(
    sentence,
    padding='max_length',
    max_length = 20,
    truncation = True,
    return_tensors = 'pt'
)
print("Input ID's:    ", tokens['input_ids'])
print("Attention Mask:", tokens['attention_mask'])
print("Token Type IDs:", tokens['token_type_ids'])
```

```python
readable = tokenizer.convert_ids_to_tokens(tokens['input_ids'][0])
print(readable)
```

```python
sentence2 = "The disinformation campaign was unprecedented."

tokens2 = tokenizer.tokenize(sentence2)
print(tokens2)
```

```python
model = BertModel.from_pretrained('bert-base-uncased')
model.eval()
with torch.no_grad():
    output = model(**tokens)

print("Last hidden state shape:", output.last_hidden_state.shape)
print("CLS token output shape: ", output.last_hidden_state[:, 0, :].shape)
```
